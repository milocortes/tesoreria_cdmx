# IMPUESTO SOBRE NÓMINAS
(Artículos 156 al 159 del CFCDMX)

Sujeto: Las personas físicas y morales que, en la Ciudad de México, realicenn erogaciones en dinero o en especie por concepto de remuneración al trabajo personal subordinado, independientemente de la designación que se les otorgue.

Objeto:
Las erogaciones destinadas a remunerar el trabajo personal subordinado: sueldos y salarios; tiempo extraordinario de trabajo; premios, primas, bonos, estímulos e incentivos; compensaciones; gratificaciones y aguinaldos; participación patronal al fondo de ahorros; primas de antigüedad; comisiones, y pagos realizados a administradores, comisarios o miembros de los consejos directivos de vigilancia o administración de sociedades y asociaciones.
Base:
El monto total de las erogaciones realizadas por concepto de remuneraciones al trabajo personal subordinado.

Tasa: 3%.

Epoca de pago: Al momento en que se realicen las erogaciones y a más tardar a los
17 días del mes siguiente.

Reducciones: Artículos 276, 277, 278, 279 y 283

```python
df_delegacion=pd.read_excel('diccionario_de_datos_1.xlsx',sheet_name='delegación-subdelegación',skiprows=[0])
df_delegacion
```

```python
# ta_sal:=  Puestos de trabajo afiliados con un salario asociado
# teu_sal:=  Puestos de trabajo eventuales urbanos con un salario asociado
# tec_sal:=  Puestos de trabajo eventuales del campo con un salario asociado
# tpu_sal:=  Puestos de trabajo permanentes urbanos con un salario asociado
# tpc_sal:=  Puestos de trabajo permanentes del campo con un salario asociado
ta_sal_vect=[]
teu_sal_vect=[]
tec_sal_vect=[]
tpu_sal_vect=[]
tpc_sal_vect=[]
subdel_vect=[]

df_delegacion_CDMX_norte=df_delegacion[(df_delegacion['ID_DELEG_RP']==39)]
df_delegacion_CDMX_sur=df_delegacion[(df_delegacion['ID_DELEG_RP']==40)]

for cve_subdel in df_asg_2023_01_31_CDMX_sur['cve_subdelegacion'].unique():    
    subdel=df_delegacion_CDMX_sur[df_delegacion_CDMX_sur['ID_SUBDEL_RP']==cve_subdel]['descripcion subdelegación'].values[0]
    subdel_vect.append(subdel)
    ta_val=df_asg_2023_01_31_CDMX[df_asg_2023_01_31_CDMX['cve_subdelegacion']==cve_subdel]['ta_sal'].values.sum()
    ta_sal_vect.append(ta_val)
    
    
import matplotlib.pyplot as plt
plt.bar(range(len(subdel_vect)),ta_sal_vect)
plt.ylabel('ta_sal',size=20,color='w')
plt.xlabel('Sub-delegación',size=20,color='w')
plt.title('Puestos de trabajo afiliados con un salario asociado \n por Sub-delegación de CDMX Sur',size=20,color='w')
plt.xticks(range(len(subdel_vect)),subdel_vect,rotation=70,color='k' )
plt.yticks(color='k')

plt.show()

for cve_subdel in df_asg_2023_01_31_CDMX_norte['cve_subdelegacion'].unique():    
    subdel=df_delegacion_CDMX_norte[df_delegacion_CDMX_norte['ID_SUBDEL_RP']==cve_subdel]['descripcion subdelegación'].values[0]
    subdel_vect.append(subdel)
    ta_val=df_asg_2023_01_31_CDMX[df_asg_2023_01_31_CDMX['cve_subdelegacion']==cve_subdel]['ta_sal'].values.sum()
    ta_sal_vect.append(ta_val)
    
    
import matplotlib.pyplot as plt
plt.bar(range(len(subdel_vect)),ta_sal_vect)
plt.ylabel('ta_sal',size=20,color='w')
plt.xlabel('Sub-delegación',size=20,color='w')
plt.title('Puestos de trabajo afiliados con un salario asociado \n por Sub-delegación de CDMX Norte',size=20,color='w')
plt.xticks(range(len(subdel_vect)),subdel_vect,rotation=70,color='k' )
plt.yticks(color='k')
plt.show()
    
```

```python
#df_secec1['descripción sector_economico_1'].unique()

sec1_dic={0:'Agricultura, ganadería, silvicultura, pesca y caza',
          1:'Industrias extractivas',
          2:'Industrias de transformación',
          3:'Industrias de transformación',
          4:'Industria de la construcción', 5:'Industria eléctrica, y captación y suministro de agua potable',
          6:'Comercio', 
          7:'Transportes y comunicaciones',
          8:'Servicios para empresas, personas y el hogar', 
          9:'Servicios sociales y comunales'}
```

```python
sec1_vect=[]
ta_sal_vect=[]
for cve_secec1 in df_asg_2023_01_31_CDMX['sector_economico_1'].unique():
    try:
        secec1=sec1_dic[cve_secec1]
        sec1_vect.append(secec1)
        ta_val=df_asg_2023_01_31_CDMX[df_asg_2023_01_31_CDMX['sector_economico_1']==cve_secec1]['ta_sal'].values.sum()
        ta_sal_vect.append(ta_val)
#         print(cve_secec1)
#         print(secec1)
    except:
        print('fail',cve_secec1)
        pass
    
    
import matplotlib.pyplot as plt
xax=np.arange(len(ta_sal_vect))
plt.bar(xax,ta_sal_vect,width=0.5)
plt.ylabel('ta_sal',size=20,color='w')
# plt.xlabel('Sub-delegación',size=20)
plt.title('Puestos de trabajo afiliados con un salario asociado \n por sector económico 1 en CDMX',size=20,color='w')
plt.xticks(xax,sec1_vect,rotation=85,color='k' )
plt.yticks(color='k')

plt.show()
```
