```python
# Rango salarial en salarios minimos, diccionario de datos

df_rango_sal=pd.read_excel('../row_data/diccionario_de_datos_1.xlsx',sheet_name='Rango salario',skiprows=[0])
df_rango_sal['rango_salarial'].fillna('NA',inplace=True)
df_rango_sal
```

```python
# Substituir los NaN en el rango salarial por el string 'NA', 
# para poder clasificarlos más facilmente:

df_asg_2023_01_CDMX['rango_salarial'].fillna('NA',inplace=True)
df_asg_2023_01_CDMX['rango_salarial'].unique()
```

```python
# Seleccionar los datos donde el rango salarial corresponde a un indicador específico
# y a los indicadores de puestos de trabajo 

# ta:=  Puestos de trabajo afiliados al IMSS (empleos asegurados o asegurados asociados a un empleo)
# teu:=  Puestos de trabajo eventuales urbanos
# tec:=  Puestos de trabajo eventuales del campo
# tpu:=  Puestos de trabajo permanentes urbanos
# tpc:=  Puestos de trabajo permanentes del campo
# ta_sal:=  Puestos de trabajo afiliados con un salario asociado
# teu_sal:=  Puestos de trabajo eventuales urbanos con un salario asociado
# tec_sal:=  Puestos de trabajo eventuales del campo con un salario asociado
# tpu_sal:=  Puestos de trabajo permanentes urbanos con un salario asociado
# tpc_sal:=  Puestos de trabajo permanentes del campo con un salario asociado

df_rango_trabajo=df_asg_2023_01_CDMX[df_asg_2023_01_CDMX['rango_salarial']=='W9'][['rango_uma','rango_salarial',
                                                           'ta','teu',
                                                           'tec','tpu','tpc',                                                             
                                                           'ta_sal','teu_sal',
                                                           'tec_sal','tpu_sal','tpc_sal','masa_sal_ta']]
df_rango_trabajo
```

```python
#Repetir el proceso de lectura y selección para todos los datos en la carpeta de datos para CDMX
# y obtener el


puestos_trabajo=['ta','teu',
                 'tec','tpu',
                 'tpc','ta_sal',
                 'teu_sal','tec_sal',
                 'tpu_sal','tpc_sal']

df_rango_sal['rango_salarial'].fillna('NA',inplace=True)
rango_sal=[rs for rs in df_rango_sal['rango_salarial'].unique() if len(rs)<5]

rango_sal

fecha_pt_rango_sal=['Fecha']

for pt in puestos_trabajo:
    fecha_pt_rango_sal.append(pt)

for pt in puestos_trabajo:
    for rs in rango_sal:
        pt_rs=pt+'_'+rs
        fecha_pt_rango_sal.append(pt_rs)

df_fecha_pt_rs=pd.DataFrame(columns=fecha_pt_rango_sal)
df_fecha_pt_rs['Fecha']=[f[4:11] for f in asg_files]


import time

t00 = time.time()

write_dir=parentdir+'/IMSS_CDMX_Rango_Salarial/'

if not os.path.exists(write_dir):
    os.makedirs(write_dir)

for asg_data in asg_files:
    t01 = time.time()
    
    print('Procesando:',asg_data)
#     Leer dataframe
    df_asg=pd.read_csv(mypath+asg_data,low_memory=False)
    
    # Substituir los NaN en el rango salarial por el string 'NA', 
    # para poder clasificarlos más facilmente:

    df_asg['rango_salarial'].fillna('NA',inplace=True)
    
    fecha=asg_data[4:11]
    
    for rs in rango_sal:
        df_rs=df_asg[df_asg['rango_salarial']==rs]
        for pt in puestos_trabajo:
            pt_rs=pt+'_'+rs
            pt_rs_total=sum(df_rs[pt].values)
#             pt_total=sum(df_asg[df_asg['rango_salarial']=='W2']['ta'])
            
            df_fecha_pt_rs.loc[df_fecha_pt_rs['Fecha']==fecha,pt_rs]=pt_rs_total
        
    for pt in puestos_trabajo:
        pt_total=sum(df_asg[pt].values)
        
        df_fecha_pt_rs.loc[df_fecha_pt_rs['Fecha']==fecha,pt]=pt_total

    
    t11 = time.time()

    total11 = t11-t01 
    
    print(total11)

t10 = time.time()


total00 = t10-t00

for pt in puestos_trabajo:
    pt_w=pt+'_W'
    pt_na=pt+'_NA'
    pt_rs_total=pt+'_salario_total'
    selected_columns = df_fecha_pt_rs.filter(regex='^'+pt_w+'|^'+pt_na)
    
    for pt_RS in selected_columns.columns:
        
        pt_RS_frac=pt_RS+'_frac'
        
        df_fecha_pt_rs[pt_RS_frac] = df_fecha_pt_rs.apply(
        lambda row: row[pt_RS]/row[pt] if row[pt]!=0 else 0 ,axis=1)
    
df_fecha_pt_rs


print('Tiempo total=',total00)

df_fecha_pt_rs
```

```python
#Leer datos de Banxico

salario_minimo=pd.read_csv('../row_data/Salario_Minimo.csv', on_bad_lines='skip')

salario_minimo
```

```python
# Convert the 'Date' column to datetime object with format 'dd/mm/yy'
salario_minimo['Fecha'] = pd.to_datetime(salario_minimo['Fecha'], format='%d/%m/%y')

# Filter the dates within the range 1998 to 2005
salario_minimo = salario_minimo[(salario_minimo['Fecha'].dt.year >= 1998) & 
                                (salario_minimo['Fecha'].dt.year <= 2023)]

# Change the date format to 'yyyy-mm'
salario_minimo['Fecha'] = salario_minimo['Fecha'].dt.strftime('%Y-%m')
salario_minimo
```

```python
df_fecha_pt_rs_salmin=df_fecha_pt_rs.merge(salario_minimo,on='Fecha',how='right')
df_fecha_pt_rs_salmin
```
