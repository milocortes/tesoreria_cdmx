# Calcular masa salarial mensual

La masa salarial guardada en el archivo `Masa_sal_Sec_2.csv` viene por día y esta dada por tipo de puesto de trabajo: 

`{'masa_sal_ta':  'Masa salarial asociada a puestos de trabajo afiliados',
'masa_sal_teu': 'Masa salarial asociada a puestos de trabajo eventuales urbanos',
'masa_sal_tec': 'Masa salarial asociada a puestos de trabajo eventuales del campo',
'masa_sal_tpu': 'Masa salarial asociada a puestos de trabajo permanentes urbanos',
'masa_sal_tpc': 'Masa salarial asociada a puestos de trabajo permanentes del campo'}`,

y por sector económico 2 del IMSS (para más información, ver la libreta `Masa_salarial_por_sector_econ.ipynb`).

En la siguiente celda, se hace la conversión a masa salarial mensual, multiplicando la masa slarial diaria de cada sector económico por el número de días de cada mes. Este dato se guarda con el formato: tipo de masa sal. _ sector econ. 2 _ 'total mensual', por ejemplo: `masa_sal_tpc_91_total_mensual`. Al mismo tiempo, se calcula la tasa (2.5% si es anterior al 2014, 3% si es posterior) correspondiente al impuesto sobre nómina de cada sector. Este dato se guarda con el formato: tipo de masa sal. _ sector econ. 2 _ 'tasa_mensual', por ejemplo: `masa_sal_tpc_91_tasa_mensual`.

```python
# Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Leer Masa_sal_Sec_2.csv

df_fecha_ms_sec2=pd.read_csv('../IMSS_CDMX_Masa_Salarial/Masa_sal_Sec_2.csv')

# Sector economico 2 de diccionario de datos

df_sec2=pd.read_excel('../row_data/diccionario_de_datos_1.xlsx',sheet_name='sector 2',skiprows=[0])

# Hacer arreglo de nombre de masa_sal y sector econ. 2
sec2=df_sec2['sector_economico_2_2pos'].values
masa_sal_col=['masa_sal_ta', 'masa_sal_teu', 'masa_sal_tec', 'masa_sal_tpu', 'masa_sal_tpc']

# Incluir Año-Mes y cantidad de eias por mes (Días)

df_fecha_ms_sec2['Fecha_datetime'] = pd.to_datetime(df_fecha_ms_sec2['Fecha'])

df_fecha_ms_sec2['Año'] = df_fecha_ms_sec2['Fecha_datetime'].dt.year
df_fecha_ms_sec2['Mes'] = df_fecha_ms_sec2['Fecha_datetime'].dt.month

import calendar

df_fecha_ms_sec2['Días'] = df_fecha_ms_sec2.apply(lambda row: calendar.monthrange(row['Año'], row['Mes'])[1], axis=1)


# Calcular total de salarios pagados por mes y la tasa correspondiente 
# por tipo de empleo (eventual, permanente, de campo o de ciudad) y por sector economico:

for ms in masa_sal_col:
    df_fecha_ms_sec2[ms+'_total_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms]*row['Días'],axis=1)
    
    df_fecha_ms_sec2[ms+'_tasa_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                row[ms]*row['Días']*0.03 if float(row['Fecha'][0:4])>=2014
                                                else row[ms]*row['Días']*0.025 ,axis=1)
    
    for s2 in sec2:
        
        ms_s2=ms+'_'+str(s2)
        
        df_fecha_ms_sec2[ms_s2+'_total_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2]*row['Días'],axis=1)
        df_fecha_ms_sec2[ms_s2+'_tasa_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                row[ms_s2]*row['Días']*0.03 if float(row['Fecha'][0:4])>=2014
                                                else row[ms_s2]*row['Días']*0.025 ,axis=1)
df_fecha_ms_sec2
```

# Cálculo de compensaciones por sector económico

Para estimar el porcentaje de la masa salarial que corresponde a las compensaciones de cada sector económico, se utilizan los diccionarios: `cw_imss_scian`, para relacionar los sectores económicos del SCIAN con los del IMSS, y `ponderadores_censo_imss` para obtener dichos porcentajes para cada tipo de compensación:
`'p_seguridad_social','p_otras_prestaciones','p_utilidades','p_indem_liqui'`.

La metodología para obtener los porcentajes de las compensaciones, esta basada en los datos de encuestas del INEGI y se explica en una carpeta a parte, llamada ` `.

En la siguiente celda, se añaden al data set `Masa_sal_Sec_2.csv` los datos del porcentaje de compensaciones por tipo de puesto de trabajo y sector económico, añadiendo al final de cada nombre el tipo de compensación que representa. Por ejemplo, `masa_sal_ta_99_p_indem_liqui` representa la parte de la masa slarial total del sector económico 2 `99` que se va en liquidaciones.

Tembién se calculan el total diario, el total mensual y la tasa (2.5% o 3%) del total de la masa salarial más las compensaciones correpondientes a cada sector. Por ejemplo, `masa_sal_tpc_99_tasa_mensual_con_compensaciones`  representa la parte de la masa slarial de trabajadores permanentes de campo del sector económico 2 `99` más el total de compensaciones.

Y finalmente el total por tipo de puesto de trabajo más compensaciones. Por ejemplo, `masa_sal_tpc_total_mensual_con_compensaciones` representa la masa slarial de trabajadores permanentes de campo más el total de compensaciones.


```python
# Sumar compensaciones a la masa salarial de cada sector

from crosswalks_imss_scian import cw_imss_scian
from ponderadores_censo_imss import ponderadores_censo_imss
# len(ponderadores_censo_imss.keys()),len(cw_imss_scian.values()),len(set(cw_imss_scian.values()))
# set(cw_imss_scian.values())-set(ponderadores_censo_imss.keys())

for ms in masa_sal_col:
    for s2 in cw_imss_scian.keys():
        ms_s2=ms+'_'+s2
        scian=cw_imss_scian[s2]
        if scian in ponderadores_censo_imss.keys():
            
            ponderadores_dict=ponderadores_censo_imss[scian]
            
            for pond in ponderadores_dict:
                
                pond_frac=ponderadores_dict[pond]
                
                df_fecha_ms_sec2[ms_s2+'_'+pond]=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2]*pond_frac,axis=1)
                
                df_fecha_ms_sec2[ms_s2+'_'+pond+'_total_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2+'_total_mensual']*pond_frac,axis=1)
                
                df_fecha_ms_sec2[ms_s2+'_'+pond+'_tasa_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2+'_tasa_mensual']*pond_frac,axis=1)
                
        else:
            for pond in ['p_seguridad_social','p_otras_prestaciones','p_utilidades','p_indem_liqui']:
                
                pond_frac=0               
#                 if pond=='p_seguridad_social':
#                     pond_frac=0
                
                df_fecha_ms_sec2[ms_s2+'_'+pond]=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2]*pond_frac,axis=1)
                df_fecha_ms_sec2[ms_s2+'_'+pond+'_total_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2+'_total_mensual']*pond_frac,axis=1)
                
                df_fecha_ms_sec2[ms_s2+'_'+pond+'_tasa_mensual']=df_fecha_ms_sec2.apply(lambda row:
                                                            row[ms_s2+'_tasa_mensual']*pond_frac,axis=1)
                
            
        df_fecha_ms_sec2[ms_s2+'_total_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        row[ms_s2]+sum([row[ms_s2+'_'+pond] for pond in ['p_seguridad_social',
                               'p_otras_prestaciones',
                               'p_utilidades','p_indem_liqui']]),axis=1)
        
        df_fecha_ms_sec2[ms_s2+'_total_mensual_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        row[ms_s2+'_total_mensual']+sum([row[ms_s2+'_'+pond+'_total_mensual'] for pond in ['p_seguridad_social',
                               'p_otras_prestaciones',
                               'p_utilidades','p_indem_liqui']]),axis=1)
        
        df_fecha_ms_sec2[ms_s2+'_tasa_mensual_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        row[ms_s2+'_tasa_mensual']+sum([row[ms_s2+'_'+pond+'_tasa_mensual'] for pond in ['p_seguridad_social',
                               'p_otras_prestaciones',
                               'p_utilidades','p_indem_liqui']]),axis=1)
        
        
    df_fecha_ms_sec2[ms+'_total_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_total_con_compensaciones'] for s2 in cw_imss_scian.keys()]),axis=1)
    
    df_fecha_ms_sec2[ms+'_total_mensual_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_total_mensual_con_compensaciones'] for s2 in cw_imss_scian.keys()]),axis=1)
    
    df_fecha_ms_sec2[ms+'_tasa_mensual_con_compensaciones']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_tasa_mensual_con_compensaciones'] for s2 in cw_imss_scian.keys()]),axis=1)
              

df_fecha_ms_sec2
```

# Cálculo de participación de trabajadores asalariados informales, respecto a asalariados formales

Se considera a un trabajador asalariado informal si no puede ejercer derechos laborales. Estos trabajadores no estan presentes en la base de datos de IMSS pero si en los censos que realiza en INEGI. 

Para estimar la participación en la masa slarial de trabajadores asalariados informales en el mercado laboral de la CDMX, en la siguiente celda se calcula el ingreso de trabajadoes en esta condición respecto al total de ingresos registrados en el censo 2021.

Las bases de datos `salarios_enoe_IV_2021.csv` y `salarios_sector_enoe_IV_2021.csv` fue extraída de la Encuesta Nacional de Ocupación y Empleo 2021 (ENOE). La columna `mh_col` contiene dos etiquetas, la etiqueta `1` indica que los datos corresponden a trabajadores asalriados informales y la etiqueta `2` que los datos corresponden a asalariados formales. En la columna `scian` se encuentra la etiqueta del sector de SCIAN y la columna `ingocup` se indica el ingreso registrado para cada sector y tipo de asalariado.  Éstos datos deben de tener un valor `cv` menor a 0.2 para tener relevancia estadística, es por eso que en la siguiente celda se cifran de esta manera.

Para encontrar la participación por sector y por tipo de puesto de trabajo (formal o informal) se divide el ingreso de cada uno por la suma de ambos ingresos en cada sector. Para los sectores donde no existen datos estadísticamente relevantes de ambos tipos de puestos de trabajo, la participación del sector informal se toma como la participación global de este tipo de empleo.

En la siguiente celda, se calcula la fracción de la masa masa salarial mensual, multiplicando la masa slarial diaria de cada sector económico por la fraccieon correspondiente. Este dato se guarda con el formato: tipo de masa sal. _ sector econ. 2 _ '_frac_informales', por ejemplo: `masa_sal_tpc_91_frac_informales`.

Tembién se calculan el total diario, el total mensual y la tasa (2.5% o 3%) del total de la masa salarial más las compensaciones correpondientes a cada sector. Por ejemplo, `masa_sal_tpc_94_total_mensual_frac_informales`  representa la parte de la masa slarial de trabajadores permanentes de campo del sector económico 2 `94` más los salarios estimados que se pagan a trabajadoes informales.

Y finalmente el total por tipo de puesto de trabajo más compensaciones. Por ejemplo, `masa_sal_tpc_tasa_mensual_con_informales` representa tasa de la masa slarial de trabajadores permanentes de campo más los salarios estimados que se pagan a trabajadoes informales.

```python
from crosswalks_imss_scian import cw_imss_scian
from ponderadores_censo_imss import ponderadores_censo_imss

for ms in masa_sal_col:
    for s2 in cw_imss_scian.keys():
        ms_s2=ms+'_'+s2
        scian=cw_imss_scian[s2]
        
        # Participación de asalariados informales para el sector económico cuya clave empieza con'3'
        if str(scian).startswith('3'):
                
            informal_frac=357774702/869704395

            df_fecha_ms_sec2[ms_s2+'_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2]*informal_frac,axis=1)

            df_fecha_ms_sec2[ms_s2+'_total_mensual_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2+'_total_mensual']*informal_frac,axis=1)

            df_fecha_ms_sec2[ms_s2+'_tasa_mensual_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2+'_tasa_mensual']*informal_frac,axis=1)
                
        else:
            
            # Participación de asalariados informales el resto de los sectores, como el sector global
            
            informal_frac=3427188007/10455270773

            df_fecha_ms_sec2[ms_s2+'_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2]*informal_frac,axis=1)

            df_fecha_ms_sec2[ms_s2+'_total_mensual_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2+'_total_mensual']*informal_frac,axis=1)

            df_fecha_ms_sec2[ms_s2+'_tasa_mensual_frac_informales']=df_fecha_ms_sec2.apply(lambda row:
                                                        row[ms_s2+'_tasa_mensual']*informal_frac,axis=1)
        
        
    df_fecha_ms_sec2[ms+'_total_con_informales']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_frac_informales'] for s2 in cw_imss_scian.keys()]),axis=1)
    
    df_fecha_ms_sec2[ms+'_total_mensual_con_informales']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_total_mensual_frac_informales'] for s2 in cw_imss_scian.keys()]),axis=1)
    
    df_fecha_ms_sec2[ms+'_tasa_mensual_con_informales']=df_fecha_ms_sec2.apply(lambda row:
        sum([row[ms+'_'+s2+'_tasa_mensual_frac_informales'] for s2 in cw_imss_scian.keys()]),axis=1)
              

df_fecha_ms_sec2
```

# Masa salarial anual

En la siguiente celda, se calcula la masa salarial, la masa salarial con prestaciones y la masa salarial de informales por año.

```python
años=list(range(2010,2024))

df_masa_sal_anual=pd.DataFrame(años,columns=['Año'])

# Calcular masa salarial anual total

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_total_mensual'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_total_mensual': ms+'_total_anual'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_total_mensual'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_total_mensual': ms_s2+'_total_anual'})
        
# Calcular tasa anual total

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_tasa_mensual'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_tasa_mensual': ms+'_tasa_anual'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_tasa_mensual'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_tasa_mensual': ms_s2+'_tasa_anual'})
        
        
# Calcular _total_anual_con_compensaciones

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_total_mensual_con_compensaciones'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_total_mensual_con_compensaciones': ms+'_total_anual_con_compensaciones'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_total_mensual_con_compensaciones'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_total_mensual_con_compensaciones': ms_s2+'_total_anual_con_compensaciones'})

        
# Calcular _tasa_anual_con_compensaciones

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_tasa_mensual_con_compensaciones'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_tasa_mensual_con_compensaciones': ms+'_tasa_anual_con_compensaciones'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_tasa_mensual_con_compensaciones'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_tasa_mensual_con_compensaciones': ms_s2+'_tasa_anual_con_compensaciones'})

# Calcular _total_anual_con_informales

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_total_mensual_con_informales'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_total_mensual_con_informales': ms+'_total_anual_con_informales'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_total_mensual_frac_informales'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_total_mensual_frac_informales': ms_s2+'_total_anual_frac_informales'})

        
# Calcular _tasa_anual_con_informales

for ms in masa_sal_col:

    df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms+'_tasa_mensual_con_informales'].sum().reset_index(),on='Año')

# Renombrar la columna de impuestos sumados
    df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms+'_tasa_mensual_con_informales': ms+'_tasa_anual_con_informales'})

    for s2 in sec2:
        ms_s2=ms+'_'+str(s2)
        
        df_masa_sal_anual=df_masa_sal_anual.merge(df_fecha_ms_sec2.groupby('Año')[ms_s2+'_tasa_mensual_frac_informales'].sum().reset_index(),on='Año')
        # Renombrar la columna de impuestos sumados

        df_masa_sal_anual = df_masa_sal_anual.rename(columns={ms_s2+'_tasa_mensual_frac_informales': ms_s2+'_tasa_anual_frac_informales'})

df_masa_sal_anual
```
