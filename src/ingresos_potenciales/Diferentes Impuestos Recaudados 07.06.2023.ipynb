```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ingresos_anuales.csv')
df
```

```python
predial =[]
for i in range (1,44,2):
    predial.append(float(df.iloc[4,i]))
 

print(predial)
```

```python
#PREDIAL

plt.subplots(figsize=(8, 4))

Total =  [int(np.nan_to_num(float(df.iloc[2,i]))) for i in range(1,44,2)]
Predial =  [int(np.nan_to_num(float(df.iloc[4,i]))) for i in range(1,44,2)]
Adquisición_de_inmuebles =  [int(np.nan_to_num(float(df.iloc[5,i]))) for i in range(1,44,2)]
Tenencia_o_uso_de_vehículos =  [int(np.nan_to_num(float(df.iloc[6,i]))) for i in range(1,44,2)]
Espectáculos_públicos =  [int(np.nan_to_num(float(df.iloc[8,i]))) for i in range(1,44,2)]
Loterias_rifas_sorteos_concursos =  [int(np.nan_to_num(float(df.iloc[9,i]))) for i in range(1,44,2)]
Servicios_de_hospedaje =  [int(np.nan_to_num(float(df.iloc[10,i]))) for i in range(1,44,2)]
Nominas =  [int(np.nan_to_num(float(df.iloc[14,i]))) for i in range(1,44,2)]

x =['2001','2002','2003','2004','2005,','2006','2007','2008','2009', '2010', '2011','2012','2013','2014','2015,','2016','2017','2018','2019', '2020','2021','2022'] 

plt.plot(x, Total, label='Total')
plt.plot(x, Predial, label='Predial')
plt.plot(x, Adquisición_de_inmuebles, label='Adquisición_de_inmuebles')
plt.plot(x, Tenencia_o_uso_de_vehículos, label='Tenencia_o_uso_de_vehículos')
plt.plot(x, Espectáculos_públicos, label='Espectáculos_públicos')
plt.plot(x, Loterias_rifas_sorteos_concursos, label='Loterías_rifas_sorteos_concursos')
plt.plot(x, Servicios_de_hospedaje, label='Servicios_de_hospedaje')
plt.plot(x, Nominas, label='Nominas')

#plt.plot(3,df_datos.iloc[3,3], marker="*", color="red",  label='without data')
plt.xticks(rotation=45)
plt.xlabel('Año')
plt.ylabel('Millones de Pesos')
plt.title('Ingresos por Impuestos',fontsize=20)
plt.legend()
plt.show()
plt.savefig('todoxx.jpg', bbox_inches='tight', dpi=150)
```
